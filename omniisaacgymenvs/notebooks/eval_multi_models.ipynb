# Aggregate Score to rank models performance

The aggregate score is calculated as a weighted sum of various performance metrics. Each metric is given a weight (w_i) that reflects its relative importance in assessing the performance of Reinforcement Learning (RL) agents trained for the "go to position" task in a 2D plane.

```python

def get_aggregate_score_df(df):
    """
    Calculate the aggregate score for each model and return a dataframe with the aggregate scores
    """
    # Define weights as a dictionary
    weights = {
        'success_rate_0.02_m': 20,
        'success_rate_0.01_m': 10,
        'success_and_stay_within_0.15_m': 20,
        'avg_rew': 5,
        'avg_ang_vel': -20,
        'avg_action_count': -10
    }

    # Calculate the aggregate score using dictionary weights
    df['Aggregate_Score'] = sum(weights[key] * df[key] for key in weights.keys())

    # Normalize the aggregate score to a range of 0 to 100
    max_score = df['Aggregate_Score'].max()
    min_score = df['Aggregate_Score'].min()
    df['Normalized_Score'] = 100 * (df['Aggregate_Score'] - min_score) / (max_score - min_score)

    # Display the dataframe with the aggregate scores
    return df[['model', 'Aggregate_Score', 'Normalized_Score']].sort_values('Normalized_Score', ascending=False)

    
```

## Load all the results in a structured way

```python
# Loading all the evaluation results

model_folders = ['icra24', 'icra24_noise', 'icra24_Pose', 'icra24_Pose_new', 'icra24_fail']
eval_environments = ['nominal_xy', 'an_xy', 'avn_xy', 'uf_xy', 'uf_an_xy', 'uf_avn_xy']

# create a nested dictionary for the model folders
model_folders_dict = {folder: {env: [] for env in eval_environments} for folder in model_folders}
# Nominal models
for _, folder in enumerate(tqdm(model_folders)):
    for env in eval_environments:
        df = pd.read_csv("../evaluations/"+folder+"/"+env+".csv", index_col=[0])
        if (folder is 'icra24_fail') or (folder is 'icra24_Pose'):
            # merge the df with and without BB
            df_BB = pd.read_csv("../evaluations/"+folder+"/"+env+"_BB.csv", index_col=[0])
            df = pd.concat([df, df_BB])
            print(f'total number of episodes in merged df for folder {folder}: {len(df)}')
        model_folders_dict[folder][env].append(df)
```

# Env-wise ranking
## Creating Aggregate Score dataframe with performace Score normalized over all the models per single environment/

## 

```python
# Loading all the evaluation results

model_folders = ['icra24', 'icra24_noise', 'icra24_Pose', 'icra24_Pose_new', 'icra24_fail']
eval_environments = ['nominal_xy', 'an_xy', 'avn_xy', 'uf_xy', 'uf_an_xy', 'uf_avn_xy']

weights = {
    'success_rate_0.02_m': 20,
    'success_rate_0.01_m': 10,
    'success_and_stay_within_0.15_m': 20,
    'avg_rew': 5,
    'avg_ang_vel': -20,
    'avg_action_count': -10
}
# Create a dataframe to store the aggregated scores (and normalized scores) for model (column) and environment (row)

aggregated_scores_by_env = pd.DataFrame(index=eval_environments, columns=model_folders)

# iterate over all environments
for env in eval_environments:
    # Initialize the dictionary to store the aggregated scores for each model
    aggregated_scores = {folder: [] for folder in model_folders}
    # iterate over all models
    for _, folder in enumerate(tqdm(model_folders)):
        df = model_folders_dict[folder][env][0]
        # Calculate the aggregate score using dictionary weights
        aggregated_scores[folder] = sum(weights[key] * df[key] for key in weights.keys())
    # Normalize the aggregate score to a range of 0 to 100 over all models
    max_score = max([aggregated_scores[folder].max() for folder in model_folders])
    min_score = min([aggregated_scores[folder].min() for folder in model_folders])
    for folder in model_folders:
        aggregated_scores[folder] = 100 * (aggregated_scores[folder] - min_score) / (max_score - min_score)
    # Store the aggregated scores for each model in the dataframe
    aggregated_scores_by_env.loc[env] = aggregated_scores
```

### Creating the daframes to show the Max and Mean values per every set of models (columns) for each specific environment (rows)

```python
# Create a copy of aggregated_scores_by_env to store only the max scores for each environment and model
max_scores_by_env = aggregated_scores_by_env.copy()
for env in eval_environments:
    max_scores_by_env.loc[env] = max_scores_by_env.loc[env].apply(lambda x: max(x))

# Create a copy of aggregated_scores_by_env to store only the average scores for each environment and model
avg_scores_by_env = aggregated_scores_by_env.copy()
for env in eval_environments:
    avg_scores_by_env.loc[env] = avg_scores_by_env.loc[env].apply(lambda x: np.mean(x))
```

```python
# round the values of the max_scores_by_env dataframe to 2 decimal places
for env in eval_environments:
    max_scores_by_env.loc[env] = max_scores_by_env.loc[env].apply(lambda x: round(x, 2))
    avg_scores_by_env.loc[env] = avg_scores_by_env.loc[env].apply(lambda x: round(x, 2))
# Rename columns
max_scores_by_env.columns = ['XY_nominal', 'XY_noise', 'Pose', 'Pose_noise', 'XY_fail_1']
avg_scores_by_env.columns = ['XY_nominal', 'XY_noise', 'Pose', 'Pose_noise', 'XY_fail_1']
```

```python
# print max_scores_by_env to a latex table, with small and compact dimensions for the table
print(max_scores_by_env.to_latex(index=True, escape=False, column_format='|l|l|l|l|l|l|', longtable=False,
                                    caption='Maximum scores for each environment and model', label='tab:max_scores_by_env',
                                    bold_rows=True, multicolumn=True, multicolumn_format='|c|', na_rep='-'))
                                    
```

```python
max_scores_by_env
```

```python
# concatenate the two dataframes keeping the same number of columns, and add a column to indicate the type of score
scores_by_env = pd.concat([max_scores_by_env, avg_scores_by_env], axis=1)
# create sub-columns for each type of score (max and avg) for each environment, keeping the same index
scores_by_env.columns = pd.MultiIndex.from_product([['max', 'avg'], scores_by_env.columns])
scores_by_env
```
